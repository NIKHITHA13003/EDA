4.Find the country name that has given 0 rating

```python
rating=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index()
```

```python
rating=rating.rename(columns={0:"ratingcount"})
rating
```

* when the rating between 4.5 to 4.9 it was excellent=="DARK GREEN"
* when the rating between 4.0 to 4.4 it was very good=="GREEN"
* when the rating between 3.5 to 3.9 it was  good=="YELLOW"
* when the rating between 2.5 to 3.4 it was average=="ORANGE"
* when the rating between 1.8 to 2.4 it was poor==RED
* when the rating 0.0 it was not rated==White

```python
final_df[final_df["Rating color"]=="White"]
```

```python
final_df[final_df["Rating color"]=="White"].Country.value_counts()
```
